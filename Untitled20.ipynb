```python
df = pd.read_csv('/content/fifa_eda.csv.zip')
# Checking for the null values in the Data
df.isnull().sum()
```

```python
# Filling the null values in the Data
df['Club'] = df['Club'].fillna(df['Club'].mode()[0])
```

```python
# Checking again for the confirmation
df["Club"].isnull().sum()
```

```python
df["Value"] = df["Value"].fillna(df["Value"].mode()[0])
```

```python
df['Value'].isnull().sum()
```

```python
df['International Reputation'] = df['International Reputation'].fillna(df['International Reputation'].mode()[0])
```

```python
df['International Reputation'].isnull().sum()
```

```python
df['Skill Moves'] = df['Skill Moves'].fillna(df['Skill Moves'].mode()[0])
```

```python
df['Skill Moves'].isnull().sum()
```

```python
df['Contract Valid Until'] = df['Contract Valid Until'].fillna(df['Contract Valid Until'].mode()[0])
```

```python
df['Contract Valid Until'].isnull().sum()
```

```python
# Checking again to see if there is not any null values
df.isnull().sum()
```

```python
# Changing the data types of numerical columns to float
df['Age'] = df['Age'].astype(float)
```

```python
# Checking again for the confirmation
df['Age'].dtype
```

```python
# Exploratory Data Analysis
# Top 10 players
top_players = df.nlargest(10, 'Value')[['Name','Club','Value']]
top_players.sort_index(ascending=True)
```

Data Visualization

```python
sns.scatterplot(x='Overall', y='Wage', data=df)
plt.title('Wage vs Overall Rating')
x = df['Overall']
y = df['Wage']
plt.show()
```

Players with higher overall ratings generally earn higher wages, but some lower-rated players can have high wages due to star value or club reputation.


```python
sns.scatterplot(x='Age', y='Potential', data=df)
plt.title('Age vs Potential')
x = df['Age']
y = df['Potential']
plt.show()
```

Younger players tend to have higher potential growth, while older players generally have reached their maximum potential.

```python
sns.scatterplot(x='Weight', y='Potential', data=df)
plt.title('Weight vs Potential')
x = df['Weight']
y = df['Potential']
plt.show()
```

Younger players tend to have higher potential growth, while older players generally have reached their maximum potential.

```python
sns.scatterplot(x='Age', y='Overall', data=df)
plt.title('Age vs Overall')
x = df['Age']
y = df['Overall']
plt.show()
```

Overall rating generally increases with age in early years, peaks around mid-20s to late-20s, and slightly declines as players get older.

```python
sns.boxplot(x='Position', y='Wage', data=df)
plt.title('Position vs Wage')
plt.show()
```

Forwards and attacking positions generally earn higher wages than defenders and goalkeepers"
